思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣 例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('joker.jpg', cv2.IMREAD_COLOR)
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
while True:
    cv2.startWindowThread()
    cv2.imshow('joker', img)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyWindow('joker')
        break
```

## Open-cv 顯示三通道

```python
(B, G, R) = cv2.split(img)
```

```python
print("image shape:",img.shape)
print("Single channal image shape:",R.shape)
```

```python
#顯示三通道
cv2.imshow("Red", R)
cv2.imshow("Green", G)
cv2.imshow("Blue", B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 填補其他通道為0
zero_channal = np.zeros(img.shape[0:2],dtype='uint8')

cv2.imshow("Blue_3channal", cv2.merge([B,zero_channal,zero_channal]))
cv2.imshow("Green_3channal", cv2.merge([zero_channal,G,zero_channal]))
cv2.imshow("Red_3channal", cv2.merge([zero_channal,zero_channal,R]))
cv2.waitKey(0)
cv2.destroyAllWindows()

```

## matplotlib.pyplot顯示三通道

```python

plt.imshow(R,cmap=plt.cm.Reds_r)
plt.figure()
plt.imshow(G,cmap=plt.cm.Greens_r)
plt.figure()
plt.imshow(B,cmap = plt.cm.Blues_r)
plt.show()
```
